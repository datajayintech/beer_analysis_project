##  If you had to pick 3 beers to recommend using only the data in this dataset, which would you pick?

If I was to pick 3 beers,  I'd go based on the mean of overall reviews, taste reviews, review appearance and beer abv. 


```python
# create df and groupby beer name. Find the mean for overall, appearance, taste and beer abv
recommendation = beer_df.groupby('beer_name').agg({'review_overall': np.mean,
                                'review_appearance': np.mean,
                                'review_taste': np.mean,
                                'beer_abv': np.mean})

# Sort columns by descending order
top_n = recommendation.sort_values(['review_overall','review_appearance',
                                    'review_taste','beer_abv'], 
                        ascending=[False, False, False, False])
# Show top 3
top_n.head(3)
```

##### Answer: Top 3 Beers I would reccomend based on the popular mean in reviews 
      1. Campfire Joe
      2. Old Incinerator Brandy Barrel Barleywine
      3. Chicago Terminal Oatmeal Bourbon Stout

## 4) If I typically enjoy a beer due to its aroma and appearance, which beer style should I try?

My approach to this will be to count the number of reviews for both columns, get the mean of reviews for both columns and get the max amount of reviews for both columns. This will help determine which is popular in both areas.

```python
# Create df for the columns aroma, appearance and beer style.
beerstyle = beer_df[['review_aroma','review_appearance','beer_style']]
beerstyle.head()
```

```python
# Get the mean, count and max for a aroma,appearance and groupby beerstyle

beerstyle1 = beer_df.groupby('beer_style').agg({'review_aroma': np.mean,
                                'review_appearance': np.mean})
beerstyle1 = pd.DataFrame(beerstyle1.reset_index())
beerstyle1.columns = ['beer_style','review_aroma_mean','review_appearance_mean']

beerstyle2 = beer_df.groupby('beer_style').agg({'review_aroma': np.max,
                                             'review_appearance': np.max})
beerstyle2 = pd.DataFrame(beerstyle2.reset_index())
beerstyle2.columns = ['beer_style','review_aroma_max','review_appearance_max']

beerstyle3 = beer_df.groupby('beer_style').agg({'review_aroma': np.size,
                                             'review_appearance': np.size})
beerstyle3 = pd.DataFrame(beerstyle3.reset_index())
beerstyle3.columns = ['beer_style','review_aroma_count','review_appearance_count']

beerstyle = pd.merge(pd.merge(beerstyle2, beerstyle1, on='beer_style'),beerstyle3, on='beer_style')
```

```python
# sort columns in desc order and limit 3 for top 3
recommended = beerstyle.sort_values(['review_aroma_count','review_appearance_count','review_aroma_mean',
                                     'review_appearance_mean','review_aroma_max','review_appearance_max'], 
                                    ascending=[False, False,False, False, False, False])
top = recommended [:3]['beer_style']
top
```

##### The Top 3 Beerstyles I would reccomend are 
      1. American IPA
      2. American Double / Imperial IPA
      3. American Pale Ale (APA)
